SORTING

```python
#Sorts the data by the amount column from smallest → largest.
df.sort_values("artistpopularity")
# Solves: provies data in order on specific column 
# Used when: we me need to see data in specific order of a column 
```

```python
#Sorts the data by amount from largest → smallest.
df.sort_values("artistpopularity",ascending=False).head()
# Solves: proveds data in large to small format 
# Used when: see data in dig to small 
```

```python
#sort multiple values 
df.sort_values(["tracknumber","artistpopularity"],ascending=[True,False]).head()

# Solves:sorts data based on multiple columns in order thay have 
# Used when: have multiple column to be sorted 
```

GROUPBY


```python
df.groupby("artistname")["albumname"].count()
# Question answered: tell the total number of alboums per artists 
# Business value: to unserdatnd how meany abloumns been relesed my artist 
```

```python
df.groupby("albumname")["albumtotaltracks"].sum()

# Question answered: whar are the total number of tracks for all ablumns 
# Business value: helps to understand how meany tracks we ahve per alboumn 
```

```python
#Multiple Aggregations
df.groupby("artistname")["tracknumber"].agg(["count","min"])

# Question answered: what are total,min tracks of a artist

# Business value: to unerstand number of tracks and min number of tracks 
```

```python
#Converts the groupby result back to a normal DataFrame instead of a grouped index.
# Makes the result easier to save or visualize.
summery = df.groupby("trackname")["tracknumber"].sum().reset_index()
summery


```
