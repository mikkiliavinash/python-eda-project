```python
import pandas as pd

df = pd.read_csv("../data/cleaned_data.csv")
```

```python
list1 = df.info()
```

```python
#filter >

df[df["artistpopularity"]>40]
```

```python
#AND = & ---- Both conditions must be TRUE
df[ (df["artistpopularity"]>=100) &(df["tracknumber"] == 11) ]
```

```python
#OR = | Either condition TRUE
 
df[(df["albumtotaltracks"] ==1) | (df["albumtotaltracks"] ==3) ]
```

```python
#filter using multiple values
df[df["trackname"].isin(["OMG!","Iris"]) & (df["albumtotaltracks"] ==3)]
```

```python
#text filter - worls similer to "Like" in SQL
df[df["trackname"].str.contains("IRIS",case=False)]
```

```python
df[
    df["trackname"].isin(
        ["Iris"])]
```

SORTING

```python
#Sorts the data by the amount column from smallest → largest.
df.sort_values("artistpopularity")
```

```python
#Sorts the data by amount from largest → smallest.
df.sort_values("artistpopularity",ascending=False)
```

```python
#sort multiple values 
df.sort_values(["tracknumber","artistpopularity"],ascending=[True,False])
```

GROUPBY


```python
df.groupby("trackname")["tracknumber"].count()
```

```python
df.groupby("trackname")["tracknumber"].sum()
```

```python
#Multiple Aggregations
df.groupby("trackname")["tracknumber"].agg(["sum","count","min"])
```

```python
#Converts the groupby result back to a normal DataFrame instead of a grouped index.
# Makes the result easier to save or visualize.
summery = df.groupby("trackname")["tracknumber"].sum().reset_index()
```

```python
summery.to_csv("../data/data2_summery.csv")
```
