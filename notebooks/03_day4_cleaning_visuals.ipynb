```python
df = df.drop(columns=["trackid.1"])
```

```python
df.head()
```

```python
# Removed rows with missing artist names as they cannot be meaningfully analyzed
df = df.dropna(subset=["artistname"])
```

```python
df.isnull().sum()
```

Histogram

```python
df["artistpopularity"].hist(bins=20)
plt.show()
```

```python
df.groupby("artistname")["albumname"].nunique()\
.sort_values(ascending=False)\
.head(10)\
.plot(kind="bar")


plt.title("Top 10 Artists by Number of Unique Albums")
plt.ylabel("Number of Alboums")
plt.xlabel("Artist Name")
plt.show()
# Shows the top 10 artists based on unique albums released
# Helps identify the most productive artists without double-counting tracks
```
