```python
#find how bad missing values are 

(df.isnull().sum()/len(df))*100
```

Divides missing count by total rows → gives percentage of missing values per column.

Multiplied by 100 to convert to percent.

Why this matters:
This tells you which columns are incomplete and how bad the missingness is.
If a column has 90% missing → maybe useless.
If it has 1–5% missing → can be cleaned.
